```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import optimize
```

# General Function for Bond Price

```python
"""
c = coupon payment (semiannual)
T = time to maturity (in years)
r = vector of yields to maturity (semiannual rates)
par = par value

"""

def bondvalue(c, t, r, par):
  return c/r +(par-(c/r))*((1+r)**(-2*t))
```

```python
price = 1200
C = 40
T = 30
PAR = 1000
R = np.linspace(0.02, 0.05, num=300)

value = bondvalue(C, T, R, PAR)
```

```python
df =pd.DataFrame({'R' : R, 'Price' : value})

plt.figure()
df.plot(x = 'R', y = 'Price',
        kind="line", figsize=(20, 10))
plt.xlabel("Yield to Maturuty")
plt.ylabel("Price")
plt.title("Yield to Maturuty Plot")
plt.show()
```
