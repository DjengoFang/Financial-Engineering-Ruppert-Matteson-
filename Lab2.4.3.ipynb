```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
iter = 9
o_price = 120
mean_year = 0.05
std_year = 0.2
trading_day = 253
mean_day = mean_year/trading_day
std_day = std_year/math.sqrt(trading_day)
```

```python
logr = np.random.normal(loc=mean_day, scale=std_day, size=trading_day)
price = 120*(np.exp(np.cumsum(logr)))
np.insert(price, 0, 120)
```

```python
plt.plot(price)
plt.ylabel('Price')
plt.xlabel('Time')
plt.show()
```

```python
df_describe['Price'] = pd.DataFrame(price)
df_describe.describe()
```

```python
df_describe['Log_Return'] = np.log(df_describe.Price) - np.log(df_describe.Price.shift(1))
df_describe.describe()
```
