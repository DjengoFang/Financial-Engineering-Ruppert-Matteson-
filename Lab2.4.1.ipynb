```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/DjengoFang/Financial-Engineering-Ruppert-Matteson-/main/datasets/Stock_Bond.csv'
data = pd.read_csv(url)
data.head(5)
```

```python
data.loc[:,["GM_AC", "F_AC"]]
```

```python
data.describe()
```

```python
data.plot(x='Date', y=['GM_AC', 'F_AC'],
        kind="line", figsize=(20, 10))
plt.xlabel("Date")
plt.ylabel("Adjusted")
plt.title("GM & Ford Price data")
plt.show()
```

```python
#GM Ford net return
data['GM_Net_Return'] = data.GM_AC.pct_change()
data['F_Net_Return'] = data.F_AC.pct_change()
data['GM_Log_Return'] = np.log(data.GM_AC) -  np.log(data.GM_AC.shift(1))
data['F_Log_Return'] = np.log(data.F_AC) -  np.log(data.F_AC.shift(1))
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
data['GM_Net_Return'].plot.hist(bins = 60)
ax1.set_xlabel("Daily returns %")
ax1.set_ylabel("Percent")
ax1.set_title("GM daily returns data")
ax1.text(-0.35,200,"Extreme Low\nreturns")
ax1.text(0.25,200,"Extreme High\nreturns")
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
data['F_Net_Return'].plot.hist(bins = 60)
ax1.set_xlabel("Daily returns %")
ax1.set_ylabel("Percent")
ax1.set_title("Ford daily returns data")
ax1.text(-0.35,200,"Extreme Low\nreturns")
ax1.text(0.25,200,"Extreme High\nreturns")
plt.show()
```

```python
data.plot(x='Date', y=['GM_Net_Return', 'F_Net_Return'],
        kind="line", figsize=(20, 10))
plt.xlabel("Date")
plt.ylabel("Net Return")
plt.title("GM & Ford Net Return data")
plt.show()
```

```python
data.plot(x='Date', y=['GM_Log_Return', 'F_Log_Return'],
        kind="line", figsize=(20, 10))
plt.xlabel("Date")
plt.ylabel("Log Return")
plt.title("GM & Ford Log Return data")
plt.show()
```

```python
#GM Ford correlartion
cor1 = data['GM_AC'].corr(data['F_AC'])
cor2 = data['GM_Net_Return'].corr(data['F_Net_Return'])
cor3 = data['GM_Log_Return'].corr(data['F_Log_Return'])
print(cor1)
print(cor2)
print(cor3)
```

```python
# Microsoft & Merck net return
data['MSFT_Net_Return'] = data.MSFT_AC.pct_change()
data['MRK_Net_Return'] = data.MRK_AC.pct_change()
data['MSFT_Log_Return'] = np.log(data.MSFT_AC) -  np.log(data.MSFT_AC.shift(1))
data['MRK_Log_Return'] = np.log(data.MRK_AC) -  np.log(data.MRK_AC.shift(1))
```
