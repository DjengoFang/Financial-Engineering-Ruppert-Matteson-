# Exercise 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import shapiro
from scipy.stats import t
from scipy.stats import norm
from scipy.stats import gaussian_kde
import statsmodels.api as sm
import pylab as py
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/DjengoFang/Financial-Engineering-Ruppert-Matteson-/main/datasets/ford.csv'
data = pd.read_csv(url)
data.head(5)
```

```python
data.plot(x="X.m..d..y", y="FORD", figsize=(30,10))
```

# (a)

```python
data['FORD'].describe()
```

# (b)

```python
fig, ax = plt.subplots(figsize=(16, 8))
sns.kdeplot(data = data, x = "FORD", fill = True)
```

# (c)

```python
print(shapiro(data['FORD']))
```

# (d)

```python
for i in [1, 3, 4, 6, 10, 20, 30]:
  tdis = t.rvs(df = i, size = data.shape[0])
  sm.qqplot_2samples(100*data['FORD'], tdis, line ='45')
  plt.xlabel("Return")
  plt.ylabel("t-plots df="+str(i))
plt.show
```
